```python
import random
p1=0
p2=0
p3=0
snake_bite={17:7,54:34,62:18,64:60,87:36,93:73,95:75,98:79}
ladder={1:38,4:14,9:31,21:42,28:84,51:67,72:91,80:99}
print("Welcome to the Snake and Ladder Game")
player1=input("Name of First Player: ")
player2=input("Name of Second Player: ")
player3=input("Name of Third Player: ")

def roll_again(position,player):
    dice=random.randint(1,6)
    if dice==6:
        position=position+dice
        print("Number rolled by {}: {}".format(player,dice))
    else:
        position=position+dice
        print("Number rolled by {}: {}".format(player,dice))
    return position

def snake_ladder(position,player):
    dice=random.randint(1,6)
    print("Number rolled by {}: {}".format(player,dice))

    # Check if they rolled 6, then need to roll again
    if dice==6 and position<94:
        position=roll_again(position,player)
        
    # Check if position is beyond 100
    if position+dice>100:
        print("Oops Number rolled by {} is big, number needed is {}".format(player,100-position))
        return position
    else:
        position=position+dice
        
    # If player lands at base of the ladder
    if position in ladder:
        print("LADDER ARRIVED")
        position=ladder[position]
        position=roll_again(position,player)
        print("{} is now at: {}".format(player,position))
        
    # If player lands at the mouth of the snake
    elif position in snake_bite:
        print("OH NO SNAKE")
        position=snake_bite[position]
        print("{} is now at: {}".format(player,position))
    else:
        print("{} is now at: {}".format(player,position))
    
    return position

# Loop stops when any one player reaches 100
while p1<=100 and p2<=100 and p3<=100:
    p1=snake_ladder(p1,player1)
    if p1==100:
        print("GAME OVER {} WINS.".format(player1))
        break
    p2=snake_ladder(p2,player2)
    if p2==100:
        print("GAME OVER {} WINS.".format(player2))
        break
    p3=snake_ladder(p3,player3)
    if p3==100:
        print("GAME OVER {} WINS.".format(player3))
        break
```
